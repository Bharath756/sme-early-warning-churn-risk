# Early Warning Churn Risk Definition

## Business Context
Customer churn often becomes visible only when it is too late to intervene.
An early warning system aims to identify customers at high risk of churn
'before' the actual churn event occurs, enabling proactive retention actions.

## What is Early Warning Churn?

In this project, early warning churn is defined as identifying customers
who are likely to churn in the near future (e.g., within the next 30â€“60 days),
based on current behavior, tenure, pricing, and service engagement.

Due to the absence of explicit time-series data, this project focuses on
'relative churn risk scoring' rather than exact churn timing.

## Labeling Strategy

The churn label available in the dataset represents whether a customer
eventually churned.

This label is used as a proxy for near-term churn risk, under the assumption
that customers who churn exhibit detectable risk patterns prior to exit.

The objective is not to predict the exact churn date, but to rank customers
by relative churn risk for prioritization.

## Scope and Limitations

This project does not:
- Predict exact churn timing
- Use sequential or event-level time-series data
- Model customer behavior post-churn

Instead, it focuses on building a practical early warning risk score
using currently available customer attributes.

## Expected Output

The early warning churn system will produce:
- A churn risk score for each customer
- A ranking of customers by risk level
- A framework to combine churn risk with CLV for intervention prioritization
